```python
import seaborn as sns
import pandas as pd
```

```python
mpg=sns.load_dataset('mpg')
mpg.tail(3)
```

```python
# group by - cylinders별 마력의 평균
    # name별 연식의 평균
    # name이 공백을 기준으로 제조사[0] - 모델[나머지]
    # 제조사 컬럼을 만들고 제조사를 추가
# 필터 - mpg 15 이상인 모델
# iloc, loc를 이용해서 데이터를 row, col 다양하게 출력
# index를 재구성 - origin
mpg.groupby("cylinders",observed=True)['horsepower'].mean()
mpg.groupby("name",observed=True)['model_year'].mean()
mpg['manufacturer']=mpg['name'].apply(lambda x: x.split(' ')[0])
mpg.head(3)
mpg['manufacturer'].value_counts()

mpg[mpg['mpg']>=15].head(3)

mpg.iloc[0:3,0:4]
mpg.loc[0:3,['mpg','cylinders','horsepower']]
mpg.set_index('origin',inplace=True)
mpg.head(3)
mpg.reset_index(inplace=True)
mpg.head(3)
```

```python
mpg.groupby('cylinders')['horsepower'].mean()
mpg.groupby('name')['model_year'].mean()
```
