```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# # titanic 데이터셋 로딩
titanic = sns.load_dataset('titanic').rename(columns={'sex' : 'gender'})

# 행 인덱스 0~9 범위에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
df = titanic.loc[0:9, ['age','fare']]

# 데이터프레임 출력
titanic.head(3)
titanic_simple = titanic.drop(columns=['alive','class','who','embark_town','deck'])
titanic_simple.head(6)

# 나이는 연속형 데이터 즉, 숫자들의 집합을 연속형 데이터라고 한다.
# 생존여부 0 = 살아있다, 1 = 죽었다
# 범주형 데이터 즉, ~중에 하나 예를들어 생존여부, pclass 등
```

```python
# 불린 인덱싱(20세 미만의 조건을 충족하는 행을 필터링)
titanic_under20 = titanic[titanic['age'] < 20]
# 여성 또는 남성의 데이터만 출력
titanic_under20 = titanic_under20.rename(columns={'sex' : 'gender'})
titanic_under20[titanic_under20['gender'] == 'female']
```

```python
# 불린 인덱싱 (20세 미만의 조건을 충족하는 행을 필터링)
df[df['age'] < 20]
```

```python
# 불린 인덱싱 (loc 인덱서 활용)
df.loc[df['age'] < 20]
```

```python
# 불린 인덱싱 (NOT 논리연산)
df.loc[~(df['age'] < 20)]       #not 이라는 뜻
```

```python
# 나이가 10대(10~19세)인 승객 (AND 논리연산)
mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic[mask1]
df_teenage.head()
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객 (AND 논리연산, loc 인덱서)
mask2 = (titanic.age < 10) & (titanic.gender == 'female')
df_female_under10 = titanic.loc[mask2]
df_female_under10.head()
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객 (OR 논리연산)
# age, sex, alone 열만 선택 (loc 인덱서, fancy indexing)
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60.head()
```

```python
# 불린 시리즈를 새로운 열로 추가
df['age_under_20'] = df['age'] < 20
df.head()
```
